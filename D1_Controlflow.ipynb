# Write a program that checks if a number is prime

````python
n=int(input("Enter your Number: "))
ls=[]
for i in range(1,n+1):
  if n%i==0:
    ls.append(i)
if len(ls)>2:
  print("not prime")
else:
  print("prime")
````
