Pandas - loc:
Given the following DataFrame `df`, write a command to select all rows where the value in the 'Age'
column is greater than 30 using `loc`.

```python
import pandas as pd
```

```python
data = {'Name': ['John', 'Anna', 'Peter', 'Linda','lucky','Sadvika','Nikhil','Krishna','Pranith'],
 'Age': [28, 34, 29, 30,23,25,39,29,32],
 'City': ['New York', 'Paris', 'Berlin', 'London','Adilabad','New York', 'Paris', 'Berlin', 'London']}
df = pd.DataFrame(data)
df
```
