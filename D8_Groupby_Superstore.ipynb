13. Find the loss-making products that have been ordered more than 5 times.

```python
dt=data[['Product Name','Quantity ordered new']].groupby('Product Name').sum().reset_index()
dt
```

```python
pr=data[['Product Name','Profit']].groupby('Product Name').sum().reset_index()
pr
```

```python
new=pd.merge(dt,pr,on='Product Name')
new
```

```python
d=new[["Product Name",'Quantity ordered new','Profit']].groupby('Product Name').sum().sort_values(ascending=[True,True],by=['Quantity ordered new','Profit'])
d[(d["Quantity ordered new"]>5) & (d["Profit"]<0)]
```
