17. Group by `Category` and `Sub-Category`, then calculate the percentage contribution of each
`Sub-Category` to the total sales of its `Category`.


```python
dt=data[["Product Category","Product Sub-Category","Sales"]].groupby(["Product Category","Product Sub-Category"]).sum().reset_index()
PC=data[["Product Category","Sales"]].groupby("Product Category").sum().reset_index().rename(columns ={"Sales":"CatSales"})
ds=pd.merge(dt,PC,on="Product Category")
ds["Per"]=ds["Sales"]*100/ds["CatSales"]
ds
```

18. For each `Region` and `Segment`, find the `Sub-Category` with the highest average profit.


```python
d=data[["Region","Customer Segment","Product Sub-Category","Profit"]].groupby(["Region","Customer Segment","Product Sub-Category"]).mean("Profit").reset_index().sort_values(by=["Region","Customer Segment","Profit"],ascending=[True,True,False]).groupby("Profit").head(1)
d
```

19. Determine the total profit for each `State`, then find the `State` with the maximum total profit in
each `Region`.


```python
TS=data[["Region","State or Province","Profit"]].groupby(["Region","State or Province"]).sum().sort_values(by=["Region","State or Province","Profit"],ascending=[True,True,False]).groupby("Region").head(1)
TS
```

20. Group by `Ship Mode` and `Segment`, then calculate the total sales, and also find the percentage share of each `Ship Mode` in the total sales of each `Segment`

```python
dt=data[["Customer Segment","Ship Mode","Sales"]].groupby(["Ship Mode","Customer Segment"]).sum().reset_index()
ds=data[["Ship Mode","Sales"]].groupby("Ship Mode").sum().reset_index().rename(columns={"Sales":"ShipSales"})
dn=pd.merge(dt,ds,on="Ship Mode")
dn["Per"]=dn["Sales"]*100/dn["ShipSales"]
dn
```
