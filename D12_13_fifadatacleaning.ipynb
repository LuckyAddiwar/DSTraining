**# let us deal with Euro and * in df **

```python
df['W/F'] = df['SM'].str.replace('★','')
df['SM'] = df['SM'].str.replace('★','')
df['A/W'] = df['A/W'].str.replace('★','')
df['IR'] = df['IR'].str.replace('★','')
df['Value'] = df['Value'].str.replace('€','').str.strip()
df['Release Clause'] = df['Release Clause'].str.replace('€','').str.strip()
df['Wage'] = df['Wage'].str.replace('€','').str.strip()
```

```python
df[['Release Clause','Value', 'Wage']]
```

```python
def con_format(value):
    if 'M' in value:
        return float(value.replace('M', ''))
    elif 'K' in value:
        return float(value.replace('K', ''))/ 1000
    return float(value)
colm = ['Release Clause','Value', 'Wage']

for col in colm:
    df[col] = df[col].apply(con_format)
```

**Dealing** with / n in Hits and Team and Contract column


```python
df["Hits"].isnull().sum()
```

```python
def remove_n(value):
    value = str(value)
    if '\n' in value:
        value = value.replace('\n', '')
    if 'K' in value.upper():
        return float(value.upper().replace('K', '')) * 1000
    else:
        return float(value)

df["Hits"] = df["Hits"].apply(remove_n)
```

```python
df["Hits"].isnull().sum()
```

```python
from datetime import datetime

def format_date(date_str):
    try:
        date_obj = datetime.strptime(date_str, "%mm-%d-%Y")
        formatted_date = date_obj.strftime("%Y-%m-%d")
        return formatted_date
    except ValueError:
        return None
```

```python
df['Year'], df['Month'], df['Date']= df['Joined'].str.split(",")[1], (df['Joined'].str.split(",")[0]).str.split(" ")[0], (df['Joined'].str.split(",")[0]).str.split(" ")[1]
```

```python
result = []
for x in df['Joined']:
  date= format_date(str(x))
  result.append(date)
result
```
